### data normalization

```python
df['1980_sfs'] = df['1980']/ df['1980'].max()
```

```python
df['1980_min_max'] = (df['1980'] - df['1980'].min()) - (df['1980'].max()- df['1980'].min())
```

```python
df['1980_z_score'] = (df['1980'] - df['1980'].mean()) / df['1980'].std()
```

```python
df[['1980_sfs', '1980_min_max', '1980_z_score']].head(4)
```

```python
df.drop(labels=['1980_sfs', '1980_min_max', '1980_z_score'], axis=1, inplace = True)
```

### data binning

```python
bins = np.linspace(df['1980'].min(), df['1980'].max(), 4)
groups = ['low', 'medium', 'high']
df['1980_bin'] = pd.cut(df['1980'], bins = bins , labels= groups, include_lowest= True)
```

```python
df['1980_bin'].value_counts() 
```
